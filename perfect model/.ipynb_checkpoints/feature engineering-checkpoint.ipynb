# trabalhando com crítica de jogos x vendas

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model

import scipy
import patsy as ps
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
filepath = os.path.join("data", "Video_Games_Sales.csv")
Games = pd.read_csv(filepath)
```

```python
Games.head()
```

```python
#avaliação global sem considerar genero #tirar jogos sem avaliação critica
Games2=Games[pd.notnull(Games['Critic_Score'])]
```

```python
#tirando jogos sem dados de vendas (se existirem)
Games3=Games2[pd.notnull(Games2['Global_Sales'])]
```

### avaliando correlação de Global Sales e Critic Count

```python
X=Games3['Global_Sales']
Y=Games3['Critic_Count']
```

```python
NP_X=np.array(X)
NP_Y=np.array(Y)
```

```python
corr = np.corrcoef(NP_X[:],NP_Y[:])
print("Correlation: " + str(corr))
```

```python
#Rank Correlation - Spearman
X_rank=np.array(scipy.stats.stats.rankdata(X))
print(X_rank)
```

```python
Y_rank=np.array(scipy.stats.stats.rankdata(Y))
print(Y_rank)
```

```python
corr = np.corrcoef(X_rank[:],Y_rank[:])
print("Spearman Correlation: " + str(corr))
```

```python
#verificando se o mesmo acontece para vendas no japão
Games4=Games3[pd.notnull(Games3['JP_Sales'])]
```

```python
X=Games4['JP_Sales']
Y=Games4['Critic_Count']

NP_X=np.array(X)
NP_Y=np.array(Y)

corr = np.corrcoef(NP_X[:],NP_Y[:])
print("Correlation: " + str(corr))
```

```python
X_rank=np.array(scipy.stats.stats.rankdata(X))
print(X_rank)
```

```python
Y_rank=np.array(scipy.stats.stats.rankdata(Y))
print(Y_rank)
```

```python
corr = np.corrcoef(X_rank[:],Y_rank[:])
print("Spearman Correlation: " + str(corr))
```

```python
#verificando se algo muda para genero especifico. checando RPG
Games5=Games3[Games3['Genre']=="Role-Playing"]
```

```python
X=Games5['Global_Sales']
Y=Games5['Critic_Count']

NP_X=np.array(X)
NP_Y=np.array(Y)

corr = np.corrcoef(NP_X[:],NP_Y[:])
print("Correlation: " + str(corr))
```

```python
X_rank=np.array(scipy.stats.stats.rankdata(X))
Y_rank=np.array(scipy.stats.stats.rankdata(Y))

corr = np.corrcoef(X_rank[:],Y_rank[:])
print("Spearman Correlation: " + str(corr))
```

```python
#para RPG o resultado é diferente. vamos olhar dados mais recentes.
Games6=Games5[Games5['Year_of_Release'] > 2008]
```

```python
X=Games6['Global_Sales']
Y=Games6['Critic_Count']
NP_X=np.array(X)
NP_Y=np.array(Y)
corr = np.corrcoef(NP_X[:],NP_Y[:])
print("Correlation: " + str(corr))
```

```python
X_rank=np.array(scipy.stats.stats.rankdata(X))
Y_rank=np.array(scipy.stats.stats.rankdata(Y))

corr = np.corrcoef(X_rank[:],Y_rank[:])
print("Spearman Correlation: " + str(corr))
```

### encontrando uma boa correlação podemos ajustar uma Regressão Linear

```python
# Create linear regression object
regr = linear_model.LinearRegression()
X_train=Y.values.reshape(len(Y),1)
Y_train=X.values.reshape(len(X),1)

# Train the model using the training sets
regr.fit(X_train, Y_train)

beta1=regr.coef_
intercepto=regr.intercept_

print(beta1)
print(intercepto)
```

```python
#R quadrado 
regr.score(X_train,Y_train)
```

## usando statsmodel

```python
#GamesX = Games5

def troca(x):
    try:
        return float(x)
    except ValueError as err:
        return np.nan

Games5["User_Score"]=Games5["User_Score"].apply( troca ) 
```

```python
y,X = ps.dmatrices('Global_Sales ~ Critic_Score + User_Score +Publisher',data=Games5, return_type='dataframe')
print(type(y))
```

```python
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

## variance inflation factor

```python
# For each X, calculate VIF and save in dataframe
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
vif.round(1)
```
